```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Data untuk jumlah aturan
supports = [0.026, 0.019, 0.013, 0.006]
eclat_num_rules = [1, 4, 122, 1045]
fp_growth_num_rules = [1, 4, 122, 1045]

# Buat DataFrame untuk Seaborn
rule_data = pd.DataFrame({
    'Minimum Support': supports * 2,
    'Jumlah Aturan': eclat_num_rules + fp_growth_num_rules,
    'Algoritma': ['ECLAT'] * 4 + ['FP-Growth'] * 4
})

# Visualisasikan jumlah aturan menggunakan Seaborn sebagai diagram batang dengan label nilai absolut
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Minimum Support', y='Jumlah Aturan', hue='Algoritma', data=rule_data, palette='muted')

# Tambahkan label nilai absolut pada setiap batang
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title('Perbandingan Jumlah Aturan (ECLAT vs FP-Growth)')
plt.xlabel('Minimum Support (%)')
plt.ylabel('Jumlah Aturan')

plt.show()
```

```python
# Data
supports = [0.026, 0.019, 0.013, 0.006]
eclat_execution_time = [36.57806897163391, 95.04615616798401, 166.1016116142273, 332.7770130634308]  # in seconds
fp_growth_execution_time = [29.059352159500122, 30.23068881034851, 58.336302518844604, 60.16707181930542]  # in seconds

# Create a DataFrame for Seaborn
data = pd.DataFrame({
    'Minimum Support': supports * 2,
    'Execution Time': eclat_execution_time + fp_growth_execution_time,
    'Algoritma': ['ECLAT'] * 4 + ['FP-Growth'] * 4
})

# Visualize execution time using Seaborn as a bar plot with value labels
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Minimum Support', y='Execution Time', hue='Algoritma', data=data, palette='muted')

# Add value labels on each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title('Perbandingan Waktu Eksekusi (ECLAT vs FP-Growth)')
plt.xlabel('Minimum Support (%)')
plt.ylabel('Waktu Eksekusi (detik)')

plt.show()
```
